```python
# Basic Libraries
from matplotlib.cbook import boxplot_stats
import numpy as np
import pandas as pd
import seaborn as sb
import seaborn as sns
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
#Importing our csv file
MainData = pd.read_csv('train.csv')
MainData.head()
```

```python
#Dimension of our data
print("Data type : ", type(MainData))
print("Data dims : ", MainData.shape)
```

```python
Q1 = pd.DataFrame(MainData[['LotArea', 'GrLivArea','TotalBsmtSF', 'GarageArea', 'SalePrice']])
Q1.describe()
```

```python
f = plt.figure(figsize=(15, 4))
sb.boxplot(data=Q1['LotArea'], orient="h",  color="yellow")
quartile1 = Q1['LotArea'].quantile(0.25)
quartile3 = Q1['LotArea'].quantile(0.75)
IQR = quartile3 - quartile1 
lower = quartile1 - 1.5 * IQR
upper = quartile3 + 1.5 * IQR
outliers = Q1[(Q1['LotArea'] < lower) | (Q1['LotArea'] > upper)]
num_outliers = len(outliers)
print("Number of outliers:", num_outliers)
```

```python
f = plt.figure(figsize=(15, 4))
sb.boxplot(data=Q1['GrLivArea'], orient="h", color="yellow")
quartile1 = Q1['GrLivArea'].quantile(0.25)
quartile3 = Q1['GrLivArea'].quantile(0.75)
IQR = quartile3 - quartile1 
lower = quartile1 - 1.5 * IQR
upper = quartile3 + 1.5 * IQR
outliers = Q1[(Q1['GrLivArea'] < lower) | (Q1['GrLivArea'] > upper)]
num_outliers = len(outliers)
print("Number of outliers:", num_outliers)
```

```python
f = plt.figure(figsize=(15, 4))
sb.boxplot(data=Q1['TotalBsmtSF'], orient="h", color="yellow")
quartile1 = Q1['TotalBsmtSF'].quantile(0.25)
quartile3 = Q1['TotalBsmtSF'].quantile(0.75)
IQR = quartile3 - quartile1 
lower = quartile1 - 1.5 * IQR
upper = quartile3 + 1.5 * IQR
outliers = Q1[(Q1['TotalBsmtSF'] < lower) | (Q1['TotalBsmtSF'] > upper)]
num_outliers = len(outliers)
print("Number of outliers:", num_outliers)
```

```python
f = plt.figure(figsize=(15, 4))
sb.boxplot(data=Q1['GarageArea'], orient="h", color="yellow")
quartile1 = Q1['GarageArea'].quantile(0.25)
quartile3 = Q1['GarageArea'].quantile(0.75)
IQR = quartile3 - quartile1 
lower = quartile1 - 1.5 * IQR
upper = quartile3 + 1.5 * IQR
outliers = Q1[(Q1['GarageArea'] < lower) | (Q1['GarageArea'] > upper)]
num_outliers = len(outliers)
print("Number of outliers:", num_outliers)
```

```python
Q3A = pd.DataFrame(MainData[['GarageType', 'SalePrice']])
Q3A.head()
```

```python

Q3A["GarageType"].fillna(value = "NA", inplace = True)
Q3A_encoded = pd.get_dummies(Q3A, columns=['GarageType'])



correlation_matrix = Q3A_encoded.corr()

# Print the correlation matrix
print(correlation_matrix)
sb.heatmap(correlation_matrix.corr(), 
           linewidths = 4, annot = True, annot_kws = {"size": 10}, cmap = "YlGnBu")
```

```python
f, axes = plt.subplots(figsize = (20,15))
sb.boxplot(x="GarageType", y="SalePrice", data=Q3A)
```
