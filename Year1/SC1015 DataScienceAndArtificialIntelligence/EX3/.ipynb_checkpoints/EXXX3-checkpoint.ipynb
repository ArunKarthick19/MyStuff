```python
# Basic Libraries
from matplotlib.cbook import boxplot_stats
import numpy as np
import pandas as pd
import seaborn as sb
import seaborn as sns
import matplotlib.pyplot as plt # we only need pyplot
sb.set() # set the default Seaborn style for graphics
```

```python
#Importing our csv file
MainData = pd.read_csv('train.csv')
MainData.head()
```

```python
#Dimension of our data
print("Data type : ", type(MainData))
print("Data dims : ", MainData.shape)
```

```python
Q1A = pd.DataFrame(MainData[['LotArea', 'GrLivArea','TotalBsmtSF', 'GarageArea', 'SalePrice']])
Q1A.describe()
```

```python
f = plt.figure(figsize=(15, 4))
sb.boxplot(data = Q1A['LotArea'], orient = "h")
Q1 = Q1A['LotArea'].quantile(0.25)
Q3 = Q1A['LotArea'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = Q1A[(Q1A['LotArea'] < lower) | (Q1A['LotArea'] > upper)]
num_outliers = len(outliers)
print("Number of outliers:", num_outliers)
```

```python
f = plt.figure(figsize=(15, 4))
sb.boxplot(data = Q1A['GrLivArea'], orient = "h")
Q1 = Q1A['GrLivArea'].quantile(0.25)
Q3 = Q1A['GrLivArea'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = Q1A[(Q1A['GrLivArea'] < lower) | (Q1A['GrLivArea'] > uper)]
num_outliers = len(outliers)
print("Number of outliers:", num_outliers)
```

```python
f = plt.figure(figsize=(15, 4))
sb.boxplot(data = Q1A['GarageArea'], orient = "h")
Q1 = Q1A['GarageArea'].quantile(0.25)
Q3 = Q1A['GarageArea'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = Q1A[(Q1A['GarageArea'] < lower) | (Q1A['GarageArea'] > upper)]
num_outliers = len(outliers)
print("Number of outliers:", num_outliers)
```
